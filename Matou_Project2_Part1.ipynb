### Question 6
#### Adding Based on Existing Column

Due to the possibility that the literacy rate in decimal form would be hard to interpret after a while, I decided to add a Percentage column which would convert the long decimal into an easy readable format and another column called Literacy Health to use the Percentage Column as a basis to label the overall literacy percentage for each index.
* **BUT!!!** Before making any changes, I had to copy the DataFrame into different one titled <u>litr<u>

```python
litr = df.copy()
litr
```

```python
litr['Percentage'] = ((litr['Literacy rate'])*100).round()
litr
```

```python
def lit (num):
    if num <= 20:
        return 'Illiterate'
    if num <= 40: 
        return 'Slightly Literate'
    if num <= 60:
        return 'Average Literacy'
    if num <= 80:
        return 'Good Literacy'
    if num <= 100:
        return 'Excellent Literacy'
```

```python
litr['Literacy Health'] = litr['Percentage'].apply(lit)
litr
```
